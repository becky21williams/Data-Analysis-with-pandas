# Data Analysis of Crime Against Women in India

### Reading Crime Against Women in India data file as dataframe.
It displays crime against women in India (year-wise) :-

```python
df  = pd.read_excel('data.xlsx')
df.head(5)
```

## Data Manipulation

##### Removing "All India"

```python
c=0
for i in range(len(df)):
    if df["STATE/UT"][i]=='All India':
        c=c+1
print(c)
```

```python
df.set_index("STATE/UT", inplace = True)
```

```python
df=df.drop("All India")
df
```

## Analysis of Total cases of specific crime year-wise. 

### Input of the Crime Head is taken by the user

```python
crime=[]
head=df["CRIME HEAD"].unique()
#print("Crime heads",head)
k=0
for i in head:
    crime.append(i)
    print(f'{k}.',i)
    k+=1
#print(crime)
cr=int(input("Enter crime name from given menu(0-1): "))
print("Selected Crime:",crime[cr])
```

## Data Processing:

```python
#loop is adding crime data year-wise
cr1=[]
b=0
a=0
for z in range(2001,2013):
    for i in range(len(df)):
        if df["CRIME HEAD"][i]==crime[cr]:
            a=df[z][i]
            b+=a
    cr1.append(b)
    b=0
print(f'List of Total {crime[cr]} case registered:\n',cr1)
```

Creating x-axis...

```python
yr=['2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012']
print("List of years:\n",yr)
```

```python
yeardf =pd.DataFrame(yr)
yeardf
```

```python
crimedf =pd.DataFrame(cr1)
crimedf
```

```python
CrimeTable = pd.concat([yeardf,crimedf],axis=1)
CrimeTable.columns=["Year","Total Crime Recorded"]
CrimeTable=CrimeTable.set_index("Year")
CrimeTable
```

```python
CrimeTable=CrimeTable.reset_index("Year")
CrimeTable
```

## Analysis of increase in crime from 2001 to 2012

```python
CRH=df['CRIME HEAD'].unique()
CRH=list(CRH)
CRH.sort()
print(CRH)
```

```python
CRH
```

## Data Processing:

```python
Cgroup=df.groupby("CRIME HEAD")
Cgroup.sum()
```

## Visualization of graph comparing crimes in years 2001 and 2012

```python
plt.style.use('fivethirtyeight')
# set width of bar
barWidth = 0.25
 
# set height of bar
bars1 = []
bars2 = []
for i in range(len(Cgroup1)):
        bars1.append(Cgroup1[2001][i])
for i in range(len(Cgroup1)):
        bars2.append(Cgroup1[2012][i])
        
        
plt.figure(figsize=(15,8),)

# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

plt.title('COMPARISON OF CRIME BETWEEN 2001 AND 2012 ',fontweight='bold')

# Make the plot
plt.bar(r1, bars1, color='#73C6B6', width=barWidth, edgecolor='white', label='2001')
plt.bar(r2, bars2, color='#45B39D', width=barWidth, edgecolor='white', label='2012')


plt.ylabel('CRIME INCIDENTS',fontweight='bold')
plt.yticks(fontsize=12)


# Add xticks on the middle of the group bars
plt.xlabel('CRIME HEADS', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))],CRH ,rotation="vertical",fontsize=12)


# Create legend & Show graphic
plt.legend(fontsize=12)
plt.show()
```

## Webscrapping to fetch hex color codes from a website

```python
import bs4
import requests
colorp=[]
c=0
url="http://www.color-hex.com/popular-colors.php"
data=requests.get(url)
soup1=bs4.BeautifulSoup(data.text,"html.parser")
for z in soup1.find_all("a"):
    a1=str(z.get("title"))
    if a1[:1]=="#":
        colorp.append(a1[:7])
        c=c+1
print(c)
print(colorp)
```

```python
colorp1=set()
for i in colorp:
    colorp1.add(i)
colorp1=list(colorp1)
#print(colorp1)
colorp1=colorp1[5:39]
len(colorp1)
print(colorp1)
```
